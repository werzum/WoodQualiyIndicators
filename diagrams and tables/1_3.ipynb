```python
# Loading data
import csv
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from ... import helper_functions

# Set appropriate size of figures
sns.set(rc={'figure.figsize':(14,10)})
```

```python
table_1_3 = helper_functions.import_file("1_3.csv")
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax = sns.countplot(x="Country", data=table_1_3)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
ax.set_ylabel("Count")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax = sns.countplot(x="Region", data=table_1_3)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
ax.set_ylabel("Count")
# Add the count on top
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# One without counts

fig, ax = plt.subplots(figsize=(12, 6))
ax = sns.countplot(x="Region", data=table_1_3)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
ax.set_ylabel("Count")
plt.show()
```

```python
table_1_3_grouped = table_1_3.groupby(['Country', 'Region']).size().reset_index().pivot(columns='Country', index='Region', values=0)
table_1_3_grouped.plot(kind="bar", stacked=True)
```
