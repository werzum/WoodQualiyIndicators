```python
# Loading data
import csv
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import helper_functions

table_path = "diagrams and tables/"

table_1_3 = helper_functions.import_file(table_path+"/1_3.csv")
table_1_11 = helper_functions.import_file(table_path+"/1_11.csv")
```

```python
merged_table = pd.merge(table_1_11,table_1_3, on="ID", how="outer")
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax = sns.countplot(x="Region", data=merged_table)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
ax.set_ylabel("Count")
# Add the count on top
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# I want: count of col 2:10 grouped by
table_grouped = merged_table.groupby(['Country', 'Region']).size().reset_index().pivot(columns=merged_table.columns[2:10], index='Region', values=0)
table_grouped.plot(kind="bar", stacked=True)
```
